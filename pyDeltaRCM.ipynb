```python
import pyDeltaRCM
import matplotlib.pyplot as plt
import numpy as np
```

```python
default_delta = pyDeltaRCM.DeltaModel()
```

```python
for t0 in range(0, 200):

    default_delta.update()
        
    if np.mod(t0,50) == 0:
        plt.pcolor(default_delta.bed_elevation, vmax=-3, vmin = -8)
        plt.colorbar()
        plt.title('t = {:2d}'.format(t0))
        plt.show()
        # print(t0)

default_delta.finalize()
```
